```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
node_nums = [128,256,512]
cpu_nums = [1,4,8,16]

mean_reward_progress = np.zeros(shape=(10,len(node_nums)*len(cpu_nums)))
max_reward_progress = np.zeros(shape=(10,len(node_nums)*len(cpu_nums)))
min_reward_progress = np.zeros(shape=(10,len(node_nums)*len(cpu_nums)))

exp_num = 0
for node in node_nums:
    for cpu in cpu_nums:
        log_file = str(node)+'_'+str(cpu)+'/Training_iterations.txt'
        
        with open(log_file,'r') as f:
            lines = f.readlines()
        f.close()
        
        mean_iter_num = 0
        max_iter_num = 0
        min_iter_num = 0
        for line in lines:
            if 'episode_reward_mean' in line:
                mean_reward_progress[mean_iter_num,exp_num] = float(line.split(': ')[1])
                mean_iter_num = mean_iter_num + 1
            if 'episode_reward_max' in line:
                max_reward_progress[max_iter_num,exp_num] = float(line.split(': ')[1])
                max_iter_num = max_iter_num + 1
            if 'episode_reward_min' in line:
                min_reward_progress[min_iter_num,exp_num] = float(line.split(': ')[1])
                min_iter_num = min_iter_num + 1
        exp_num = exp_num + 1
```
